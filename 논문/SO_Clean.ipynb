```python
import spacy
import en_core_web_sm
nlp = spacy.load('en_core_web_sm')
```

```python
#데이터프레임의 데이터를 불필요한 요소 제외하고 리스트로 반환
def list_making (data) : 
    token_list = []
    #nlp에 객체 할당
    text = nlp(data)
    for token in text:
    #    if token.tag_ == "NNP" :
    #        print(True)
        #print(token.text, token.pos_, token.tag_, type(token.tag_))
    #token.pos_일 때
    #    if token.pos_ == "NOUN" or "ADJ":
    #        token_list.append(token.text)
    #token.tag_일 때
    #and : NN이 포함 안됨
    #or : 모든 요소가 다 포함됨
    #이유는 모르겠지만 따로 해야 정상적으로 list 작성된다.
    #    if token.tag_ == "NNP" :
    #        token_list.append(token.text)
    #    if token.tag_ == "NN" :
    #        token_list.append(token.text)
    #    if token.tag_ == "JJ" :
    #        token_list.append(token.text)
    #아예 처음 pos분류 이후 tag에서 골라내보기
        #if token.pos_ == "NOUN" or "PROPN": 으로 묶으면 분류가 안되는데 이유?
        if token.pos_ == "NOUN" :
            token_list.append(token.text)
        elif token.pos_ == "PROPN" :
            token_list.append(token.text)
        elif token.pos_ == "ADJ" :
            token_list.append(token.text)
        elif token.pos_ == "VERB" :
            token_list.append(token.text)
        elif token.pos_ == "ADV" :
            #장소부사일 때는 추가하지 않는다.
            if token.tag_ == "EX":
                continue
            #의문사일 때는 추가하지 않는다.
            if token.tag_ == "WRB":
                continue
            token_list.append(token.text)
    #print(token_list)
    return token_list
```

```python
#list를 string으로 변환
def listToString(str_list) :
    result = ""
    for s in str_list:
        result += s + " "
    #strip 공백제거
    return result.strip()
```

```python
#split을 쓰기 때문에 string 객체만 받아야한다.
def first_word(data) :
    blank = []
    data_list = data.split()
    for i in range(len(data_list)) :
        blank.append(data_list[i][0])
    return blank
```

```python
SO_copy = SO_raw.copy()
```

```python
for i in range(10929) :
    data = list_making(SO_copy['SO'][i])
    data = listToString(data)
    data = first_word(data)
    data = listToString(data)
    SO_copy['SO'][i] = data
    #print(SO_copy['SO'][i])
```
